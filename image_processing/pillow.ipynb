The `Image.open` method does not load image data into the computer memory. This is for lazy image loading and `PIL` implicity triggers `load` when calling certain metod.

Cropping using `variable_name[]` is like treating image like numpy array.
This method is fast since got no decoding overhead.

<img src="./src/numpy-array-slicing.webp" width="400" height="350">


```python
import os
from PIL import Image
import numpy as np
from PIL import ImageOps

baboon_path = "src/baboon.png"
cwd = os.getcwd()

baboon = os.path.join(cwd, baboon_path)

# Load the image
img = Image.open(baboon)
# Display the image
# img.show()

print(f"Image format: {img.format}")
print(f"Image size: {img.size}")
print(f"Image mode: {img.mode}")

baboon2 = img.copy()

print(f"original id: {id(img)}  copied id: {id(baboon2)}")

numpy_image = np.array(img)
print(f"Image array shape: {numpy_image.shape}")

baboon_flip = ImageOps.flip(img)
# baboon_flip.show()

# cropping a gray scale image in half
gray_array = np.array(img.convert("L"))
W, H = gray_array.shape
left_image = Image.fromarray(gray_array[:, W//2:])
left_image.show()
right_image = Image.fromarray(gray_array[:, :W//2])
right_image.show()
```
