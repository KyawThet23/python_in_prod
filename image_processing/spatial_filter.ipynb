- Spatial operations use  pixels in a neighborhood to determine the present pixel value.<br>

- Basic ideas is
  1. take a small neighborhood around each pixel
  2. apply filter (kernel)
  3. compute new pixel value

- Kernel in computer vision means samll matrix that defines how neighbours influence the center pixel.

```python
# Used to view the images
import matplotlib.pyplot as plt
# Used to perform filtering on an image
import cv2
# Used to create kernels for filtering
import numpy as np
```

```python
# Load an image
lenna_image = cv2.imread('./src/lenna.png')
```

```python
# Smoothing filter with guassian kernel using openCV
def smooth_image(image, kernel_size=5):
    smoothed_image = cv2.GaussianBlur(image, (kernel_size, kernel_size), 0)
    return smoothed_image

smoothed_image = smooth_image(lenna_image, kernel_size=7)
plt.imshow(cv2.cvtColor(smoothed_image, cv2.COLOR_BGR2RGB))
plt.title("Smoothed Image")
plt.show()
```

```python
# sharpening filter using openCV
image1 = cv2.imread('./src/mammogram.png')

def sharpen_image(image):
    kernel = np.array([[0, -1, 0],
                       [-1, 5,-1],
                       [0, -1, 0]])
    sharpened_image = cv2.filter2D(image, -1, kernel)
    return sharpened_image

sharpened_image = sharpen_image(image1)
plt.imshow(cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB))
plt.title("Sharpened Image")
plt.show()
```
